**La distancia media entre dos señales periódicas $x_1(t)$ y $x_2(t)$ pertenecientes a $\mathbb{R}$,$~\mathbb{C}$; se puede expresar a partir de la potencia media de la diferencia entre ellas:**
$$
d^2(x_1,x_2)=\bar{P}_{x_1-x_2}=\lim_{T\rightarrow\infty}\frac{1}{T}\int_{T}|x_1(t)-x_2(t)|^2dt
$$
Sea $x_1(t)$ y $x_2(t)$ dos señales como se muestra a continuación:
$$
x_1(t)=Ae^{jw_0t}\\
x_2(t)=Be^{j5w_0t}
$$
Con $w_0=\frac{2\pi}{T}$; $T,A,B \in \mathbb{R^+}$
Determine la distancia entre las dos señales.

$$d(x_1, x_2) = \bar{P}_{x_1-x_2}=\lim_{T \to \infty} \frac{1}{T} \int_T |x_1(t)-x_2(t)|^{2}~dt\\
Siendo: (a-b)^2 = a^2-2ab+b^2\\
=\int_T|x_1(t)|^2 -\int_T2(x_1(t))(x_2(t)) + \int_T|x_2(t)|^2\\
Entonces: \bar{P}_{x_1-x_2}=\lim_{T \to \infty}\left[\bar{P}_{x_1} -\frac{1}{T}\int_T2(x_1(t))(x_2(t))+ \bar{P}_{x_2}\right]\\
$$

$$Para: {x_1}(t) ~con ~T,A\in\mathbb{R^+}\\
\bar{P}_{x_1}=\frac{1}{T}\int_{T}|Ae^{jw_0t}|^2dt
$$
Observamos que $x_1(t)$ es una señal exponencial compleja, por lo que:
$$
Ae^{jw_0t}=A(cos(w_0t)+jsen(w_0t))
$$
Por lo que:
$$
\bar{P}_{x_1}=\frac{1}{T}\int_{T}|A(cos(w_0t)+jsen(w_0t))|^2dt\\
\bar{P}_{x_1}=\frac{1}{T}\int_{0}^{T}|A|^2|cos^2(w_0t)+jsen^2(w_0t)|dt
$$
teniendo en cuenta que 1 = $sen^2x$ + $cos^2x$:
$$
\bar{P}_{x_1}=\frac{1}{T}\int_{0}^{T}|A|^2dt\\
\bar{P}_{x_1}=\frac{1}{T}\left[A^2T\right]\\
\bar{P}_{x_1}=A^2
$$



$$Para: {x_2}(t) ~con ~T,B\in\mathbb{R^+}\\
\bar{P}_{x_2}=\frac{1}{T}\int_{T}|Ae^{j5w_0t}|^2dt
$$
Observamos que $x_2(t)$ es una señal exponencial compleja, por lo que:
$$
Be^{j5w_0t}=B(cos(5w_0t)+jsen(5w_0t))
$$
$$
\bar{P}_{x_2}=\frac{1}{T}\int_{T}|B(cos(5w_0t)+jsen(5w_0t))|^2dt\\
\bar{P}_{x_2}=\frac{1}{T}\int_{0}^{T}|B|^2|cos^2(5w_0t)+jsen^2(5w_0t)|dt
$$
teniendo en cuenta que 1 = $sen^2x$ + $cos^2x$:
$$
\bar{P}_{x_2}=\frac{1}{T}\int_{0}^{T}|B|^2dt\\
\bar{P}_{x_2}=\frac{1}{T}\left[B^2T\right]\\
\bar{P}_{x_2}=B^2
$$

$$
Para:-\frac{1}{T}\int_T2(x_1(t))(x_2(t)) ~con ~T,A,B \in \mathbb{R^+}\\
-\frac{2}{T}\int_{0}^{T}ABe^{jw_0t}*e^{j5w_0t}dt
$$
Teniendo en cuenta que al multiplicar exponentes son misma base se suman:
$$
-\frac{2}{T}\int_{0}^{T}ABe^{jw_0t ~+ ~5jw_0t}dt\\
-\frac{2}{T}\int_{0}^{T}ABe^{j6w_0t}dt\\
ABe^{j6w_0t}=AB(cos(6w_0t)+jsen(6w_0t))\\
-\frac{2}{T}\int_{0}^{T}AB(cos(6w_0t)+jsen(6w_0t))dt\\
-\frac{2}{T}AB\left[\int_{0}^{T}cos(6w_0t)dt+j\int_{0}^{T}sen(6w_0t)dt\right]\\
-\frac{2}{T}AB\left[\left[\frac{1}{6w_0}sen(6w_0t)\right]_{0}^{T}+j\left[-\frac{1}{6w_0}cos(6w_0t)\right]_{0}^{T}\right]\\
-\frac{2}{T}AB\left[\frac{1}{6w_0}sen(6w_0T)+j(-\frac{1}{6w_0}(cos(6w_0T)-1))\right]
$$
Reemplazando $w_0=\frac{2\pi}{T}$:
$$
-\frac{2}{T}AB\left[\frac{T}{12\pi}sen(12\pi)+j(-\frac{T}{12\pi}(cos(12\pi)-1))\right]\\
-\frac{2}{T}AB\left[\frac{T}{12\pi}0+j(-\frac{T}{12\pi}0)\right]
$$
Por lo tanto:
$-\frac{1}{T}\int_T2(x_1(t))(x_2(t))$ = 0

$$Finalizando:
\bar{P}_{x_1-x_2}=\lim_{T \to \infty} \frac{1}{T} \int_T |x_1(t)-x_2(t)|^{2}~dt=\\
\lim_{T \to \infty}\left[A^2 -0+ B^2\right]\\
\bar{P}_{x_1-x_2}=A^2+B^2
$$

#B)
**Cúal es la señal obtenida en tiempo discreto al utilizar un conversor análogo digital con frecuencia de muestreo de $5kHz$, aplicando la señal continua:**
$$x(t)=3cos(1000{\pi}t)+5sen(2000{\pi}t)+10cos(11000{\pi}t)
$$
Realizar la simulación del proceso de discretización. En caso de que la discretización no sea apropiada, diseñe e implemente un conversor adecuado para la señal estudiada.

Para ver si la discretización es apropiada debemos comprobar si cumple con el $Teorema~de~Nyquist$:
$$
F_s\geq2F_{max}
$$
Para este caso $F_s = 5000Hz$ y $F_{max} = 11000{\pi}Hz$
$$
F=\frac{w}{2\pi}\\
F_{max}=\frac{11000\pi}{2\pi}=5500Hz
$$
Verificamos $Nyquist$:
$$
5000\geq11000
$$
no se cumple la desigualdad, por lo que al haber copia no es apropiada.

Para la discretizacion:
$$
t=nT_s=\frac{n}{Fs}\\
$$
Tomando de $F_s$ el $min~ 11000Hz$:
$$
x(n)=3cos(1000{\pi}(\frac{n}{11000}))+5sen(2000{\pi}(\frac{n}{11000}))+10cos(11000{\pi}(\frac{n}{11000}))\\
x(n)=3cos(\frac{{\pi}n}{11})+5sen(\frac{2{\pi}n}{11})+10cos({\pi}n)
$$


Para ver que señal se obtenía:
$$
\Omega_{original}=\Omega_{copia}\pm2\pi\\
x(n)=3cos(\frac{{\pi}n}{5})+5sen(\frac{2{\pi}n}{5})+10cos(\frac{11{\pi}n}{5}-\frac{10}{5}\pi)\\
x(n)=3cos(\frac{{\pi}n}{5})+5sen(\frac{2{\pi}n}{5})+10cos(\frac{n{\pi}}{5})
$$


Simulando el proceso de discretización

```python
import numpy as np
import matplotlib.pyplot as plt

Fs = 11000 #Frecuencia de muestreo propuesta
Fb = 5000 #Frecuencia de muestreo anterior (para demostrar graficamente los valores perdidos)
Ti = 0 #tiempo de inicio
Tf = 2*np.pi/500 #tiempo final (2pi)
Td = np.arange(Ti, Tf, 1/Fs) #Funcion que devuelve los valores espaciados en 1/Fs entre Ti y Tf
Tb = np.arange(Ti, Tf, 1/Fb) #Funcion que devuelve los valores espaciados en 1/Fb entre Ti y Tf
Xt = 3*np.cos(1000*np.pi*Td) + 5*np.sin(2000*np.pi*Td) + 10*np.cos(11000*np.pi*Td) # la señal continua 11KHz
Xb = 3*np.cos(1000*np.pi*Tb) + 5*np.sin(2000*np.pi*Tb) + 10*np.cos(11000*np.pi*Tb) # la señal continua 5KHz

#PARTE GRAFICA
plt.figure(1, (25, 15)) #Modifica el tamaño de la figura en el display
plt.plot(Td, Xt,label='Fs=11K') #Grafica de 11kHz
plt.plot(Tb, Xb,c='r',label='Fb=5K') #Grafica de 5kHz
plt.grid() #Tablatura
plt.legend() #Titulos de cada Gráfica
plt.show()



```

Se puede observar que con el primer $Fs$ (señal roja) se pierde mucha información.

Con el nuevo $Fs$ (señal azul) se puede ver lo que el primero no tomaba.

A partir de este nuevo $Fs$, aquellos valores que sean mayores darán una mejor visibilidad a la información  

```python
#SEÑAL DISCRETA
plt.figure(2,(25,15))
plt.plot(Td, Xt,c='r',label='x(t)')
plt.stem(Td, Xt,label='Discreta x(n)') #Gráfica discreta, dibujando lineas perpendiculares al eje x
plt.grid()
plt.legend()
plt.show()
```

#C)
**Implemente una simulación para encontrar la salida del sistema lineal e invariante al tiempo $𝓗\{\cdot\}$, con respuesta al escalón $h_{\epsilon}[n]=${2,4,1,5,0,10}, $n\inℤ$ ante la entrada análoga en corriente $x(t)=20(cos\frac{t}{3}+cos(\frac{t}{4}))A~[amperios]$.**

Incluya los acondicionamientos necesarios de discretización y cuantización, asumiendo un microprocesador de 4 bits con entrada análoga de 4mA a 20mA.

Viendo $x(t)$:
$$
siendo:~ 20(cos(\omega_{1}t)+cos(\omega_{2}t))\\
\omega_{1}=\frac{1}{3},~~~\omega_{2}=\frac{1}{4}\\
T_{1}=\frac{2\pi}{\omega_1}=6\pi,~~~T_{2}=\frac{2\pi}{\omega_2}=8\pi\\
$$
Ademas:
$$
\frac{\omega_1}{\omega_2}=\frac{1/3}{1/4}=\frac{4}{3} ~el ~cual ~\inℚ
$$

Esto indica que la señal es cuasiperiodica:
$$
T=kT_1=lT_2 ~~k,l\inℤ\\
T=k6\pi=l8\pi\\
Encontrando ~MCM(6,8)=24\\
T=24\pi
$$

Para mapear la salida de $x(t)$. Generar el cero y pendiente según $(4mA;~20mA)$:
$$
y[t_n]=mx[t_n]+c\\
m=\frac{y_{max}-y_{min}}{x_{max}-x_{min}}\\
c=y[t_n]-mx[t_n]~sustituyendo~x~por~x_{min}~y~por~y_{min}
$$

```python
def open(x,ym = 0.004,yx = 0.02): #creamos una función propia para este caso, asi como ajustarle valores predeterminados de acuerdo al ejercicio actual
  '''
  Cero y pendiente
  regresa y[tn]
  '''
  mx = x.max() #para encontrar el valor maximo en la funcion
  mn = x.min() #para encontrar el valor minimo en la funcion
  m = (yx-ym)/(mx-mn) #para encontrar la pendiente
  c = ym - m*mn
  y = m*x + c #reemplazando
  return y

```

$$
siendo~asi: ~F_s\geq2F_{max}\\
F_1=\frac{\omega_1}{2\pi},~F_2=\frac{\omega_2}{2\pi}\\
F_1=\frac{1}{6\pi},~F_2=\frac{1}{8\pi}
$$
Entonces siempre que $F_s\geq\frac{1}{3\pi}$ será una frecuencia adecuada

para discretizar:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

T1 = 6*np.pi #periodo de cos(t/3)
T2 = 8*np.pi #periodo de cos(t/4)
F1 = 1/T1 #frecuencia de cos(t/3)
F2 = 1/T2 #frecuencia de cos(t/4)
Tt = 24*np.pi #periodo de (cos(t/3)+cos(t/4))
Fs = 15*F1 #frecuencia de muestreo, en este caso (1/6pi)*15
t = np.arange(0,4*Tt,1/Fs) #Funcion que devuelve los valores espaciados en 1/Fs entre Ti(para este caso 0) y Tf(4 periodos), vector del tiempo
x = 20*(np.cos(t/3)+np.cos(t/4)) #señal en cuestion
y = open(x,0.004,0.02).reshape(-1,1) #funcion previa de cero y pendiente, junto con el reorganizado del vector
```

cuantización:
$$
4bits;~2^4=16~estados
$$

```python
Stats = 16 #No de estados en base a la cantidad de bits
ymax = 0.02
ymin = 0.004
res = (ymax-ymin)/Stats #20mA-4mA /2^4 resolucion
cnt = np.arange(ymin,ymax,res).reshape(-1,1)

xcnt = np.zeros(len(y)) #nuevo vector llenado con ceros en tamaño de y
for i in range(len(y)):
  xcnt[i] = cnt[np.argmin(cdist(y[i].reshape(-1,1),cnt))] #distancia entre cada punto al cnt
```

```python
plt.figure(1,(15,10))
plt.plot(t,x,label='x(t)')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(2,(15,10))
plt.plot(t,y,c='r',label='y(t)')
plt.stem(t,xcnt,label='y[tn]')
plt.xlabel('Muestra')
plt.ylabel('Amplitud')
plt.legend()
plt.show()
```

Convolución:
$$
y[n]=x[n] * h[n]=\sum_{k=-\infty}^{\infty}x[k]h[n-k]
$$
para este caso concreto en el programa $y=Y$, $h=y$

```python
yu = np.array([2,4,1,5,0,10]) #arreglo para el escalo Hn
y = np.diff(yu)
Y = np.convolve(xcnt,y,mode='same')
plt.figure(3,(15,10))
plt.stem(t,Y,label='Y[n]')
plt.legend()
plt.xlabel('Muestra')
plt.ylabel('Amplitud')
plt.show()
```
